## **Numerical Summary Measures**

### **Measures of Central Tendency**


**Mean (Arithmatic average)**
- Most commonly used measures of central tendency
- There are different types of mean. In general we talk about **arithmatic mean**, also known as **average**.


\begin{equation}
Mean =\sum_{i=1}^n \frac {Xi}{n} 
\end{equation}

- Not appropriate for nominal or ordinal level data
- Exception: when we have dichotomous data and the two possible outcomes are represented by 0 and 1. In this situation means of the observation is equal to the proportion of 1s in the data set.
- **Extremely sensitive to unusual values, such as outliers.**


|ID|Age|
|---|---|
|1|11|
|2|12|
|3|13|
|4|14|
|5|15|

- n = 5
- Mean value will be (11 + 12 + 13 + 14 + 15)/5
- mean = 13

**Median (50th percentile)**

- When n is **odd**, we take middle value after rearrainging data in a ascending or descending order.
\begin{equation}
\frac {n + 1}{2}
\end{equation}
- The above mentioned formula gives the position of middle value.

|ID|Age|
|---|---|
|1|11|
|2|12|
|3|13|
|4|14|
|5|90|

- Here n is odd and we have an **extreme value 90**.
- If we use mean, value will be influenced by 90. Hence **mean = 28**. Influenced by 90.
- In this case median will be better.
- As n is odd so we have to use (n + 1)/2 to get the position of median value.
- Here position is 3rd, so median value will be 13

- When n is **even**, we take the average of two middle values after rearrainging data in an ascending or descending order.
\begin{equation}
average(\frac {n}{2}, \frac {n}{2} \ + 1)
\end{equation}
- The above mentioned formula gives the position of two middle most values. 
- We take the average of these values.

|ID|Age|
|---|---|
|1|10|
|2|11|
|3|12|
|4|13|
|5|14|
|6|90|

- Here n 6, so it is even.
- So 6/2 = 3 and 6/2 + 1 = 4. Average of 3rd and 4th value. 
- Median is 12.5.

### **Bar plot**

- Very commonly used to examine the distribution of categorical data. Basically tells the count.
- It is also used to see the difference between numerical and categorical data.

```python
#@title
# Counting total numbers in each category
count = mtcars['am'].value_counts()
# Checking the numbers
count
# Creating categories based on numbers
AM = ['Automatic', 'Manual']
# Creating plot
fig = plt.figure(figsize=(10,5))
plt.bar(AM, count, color=['cyan', 'pink'])
plt.title("Total number of automatic and manual")
plt.xlabel("Automatic vs Manual")
plt.ylabel("Count")
# show plot
plt.show()
```

### **Pie Chart**
- Very similar to bar plot, but used to see the percentage in each category
- Used for categorical values

```python
#@title
# Counting total numbers in each category
count = mtcars['am'].value_counts()
# Checking the numbers
count
# Creating categories based on numbers
AM = ['Automatic', 'Manual']
# Creating plot
fig = plt.figure(figsize =(10, 7))
# Show plot
plt.pie(count, labels = AM,autopct='%1.1f%%',shadow=True, startangle=90) # autopct to display the percent value using Python string formatting. For example, autopct='%1.1f%%' means that for each pie wedge, the format string is '1.1f%'
plt.title("Pie Chart")
plt.show()
```

Automatic cars are higher in percentage in comparions to manula cars.
